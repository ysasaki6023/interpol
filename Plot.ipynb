```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dList = []
d = pd.read_csv("data/data.csv")
dList.append(d)

for iIter in [1000,3000]:
    d = pd.read_csv("save/data_%d.csv"%iIter)
    dList.append(d)

fig = plt.figure()
for d in dList[::-1]:
    plt.plot(d["idx"],d["col1"])
fig.show()

fig = plt.figure()
for d in dList[::-1]:
    plt.plot(d["idx"],d["col2"])
fig.show()
```
